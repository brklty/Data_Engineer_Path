```python
import findspark
```

```python
findspark.init("/opt/manual/spark/")
```

```python
from pyspark.sql import SparkSession, functions as F
from pyspark.sql.types import *
```

```python
spark = (SparkSession.builder
        .appName("udf")
        .master("local[2]")
         .getOrCreate()     
        
        ) 
```

```python
df = spark.read \
.option("header", True) \
.option("inferSchema", True) \
.option("compression", "gzip") \
.csv("file:///home/train/datasets/Hotel_Reviews.csv.gz")
```

```python
df2 = df.withColumn("Tags", 
                    F.split(F.col("Tags"),",")
                   .cast(ArrayType(StringType()))) \
.withColumn("Review_Date",F.to_date(F.col("Review_Date"),"M/d/yyyy"))
```

```python
df2.limit(2).toPandas()
```

```python
# col rename

df3 = df2.withColumnRenamed("Additional_Number_of_Scoring", "ANOS")
```

```python
df3.limit(2).toPandas()
```

```python
# add col

df4 = df2.withColumn("Reviewer_Nationality_New", F.col("Reviewer_Nationality"))
```

```python
df4.limit(2).toPandas()
```

```python
df5 = df2.withColumn("ANOS_new", F.expr("Additional_Number_of_Scoring * 2"))
```

```python
df5.limit(2).toPandas()
```

```python
#drop col

df6 = df5.drop("ANOS_new", "Reviewer_Nationality_new")
```

```python
df6.limit(2).toPandas()
```
